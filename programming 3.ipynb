```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1: Graphing

```python
# Let's make some random data first

# We use np.arange to make random data

x = np.arange(1, 101, 1) # Function np.arange(from 1, to 101, skip 1)
y = np.arange(1, 1001, 10)/3 # Some random data

# Let's see the data
print(x); print(y)
```

```python
# Now we want to graph them

plt.plot(x, y,)
plt.show()
```

```python
plt.plot(x, y, color= "Red", linestyle = "--")

# Now we can label the graph
plt.xlabel("Number of Days") 
plt.ylabel("Height")
plt.title("Plant's height")

plt.show()
```

# 2: Scatter plot

```python
# Let's create some random numbers using numpy built in function

# Random integer

a = np.random.randint(1, 400, 100) # (From 1, to 400, 100 of them)

# Random decimals

b = np.random.rand(100) # Hundreds of them

# Let's see them

print(a); print(b)
```

```python
plt.scatter(x, y, color = "red", label = "My Spending") # We can label them and show the legend in thr graph
plt.scatter(x, a, color = "blue", label = "Other's Spending")

plt.title("Daily spending")
plt.xlabel("Days")
plt.ylabel("Dollar")
plt.legend()
plt.show()

# The graph doesn't make sense, but the key here is to learn how to plot
```

# 3: Histogram

```python
# Let's create some random integer

c = np.random.randint(50, 200, 365)
d = np.random.randint(30, 250, 365)

# And plot them using histogram function

plt.hist(c, bins = 50, color = "blue", label = "Data A")       # Bins are the intervals
plt.hist(d, bins = 50,  color = "orange", label = "Data B")
plt.legend()
plt.show()
```

```python
# There is intercept in the histogram that we cannot see clearly, it is blacked out by the other data on the intercept.
# To see that clearly, we adjust the transparency

plt.hist(c, bins = 50, alpha = 0.5, color = "blue", label = "Data A")  # Alpha = transparency (0 = transparent)
plt.hist(d, bins = 50, alpha = 0.5, color = "orange", label = "Data B")
plt.legend()
plt.show()
```

# 4: Bar chart

```python
# Having these two data, let's plot a bar chart

tickers = ["AAPL", "MSFT","GOOGL"]
values = [459.63, 208.90, 1504.63]

# We can plot the bar chart using plt.figure(figsize =(length, height), dpi= )
plt.figure(figsize =(8, 6), dpi= 300)


# We can color the bars using plt.bar function(a, b, color = "color")
bars = plt.bar(tickers, values, color = ("red", "green", "blue"))

patterns = ["/" , "o", "*"]

# This is a function to combine bars and patterns we just made:
for bar1 in bars: 
    bar1.set_hatch(patterns.pop(0))

plt.show()
```
