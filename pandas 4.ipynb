```python
import pandas as pd
```

```python
accidents = pd.read_csv("accidents.csv", encoding = "ISO-8859-1")
```

```python
# Looking at this message above, there is this warning: Columns (6) have mixed types
# let's examine this

accidents.head()
```

```python
# Let's look at our columns first

accidents.columns

# Look, there is spaces infront of each column names, we need to remove it
```

```python
# Now, we want to remove the whitespaces infront of the Column names, we do this by using .strip() function and
# Uses lambda method to apply it to all columns name

# We uses .rename() function to rename our column and input the renamed input in the function .rename()

accidents.rename(columns = lambda x: x.strip(), inplace = True)

# We use inplace = True to overide existing data frame

# Lambda x, means make accidents = to x and perform x.strip()
```

```python
accidents.columns
```

```python
# Let's observe our dataframe description

accidents.describe()
```

```python
# If you want to describe a specific column, this is how you should do

accidents["County Name"].describe()
```

```python
# The above description suggests:
# We have total 1,048,575 inputs and ther are 21 unique country input, 
# County that show up the most is MIDDLESEX, with 115,760 of it

# Now, let's see what are the 21 inputs in County Name

accidents["County Name"].unique()

# These are the 21 unique County Names 
```

```python
# Again, there is whitespaces in these inputs, we want to remove them.
# This time we'll use the map() function which will strip the white space out of every string found in the column

accidents["County Name"] = accidents["County Name"].map(str.strip) 
accidents["County Name"].unique()

# DONE!
```

```python
# Now, let's examine the data types in each column

accidents.dtypes
```

```python
# Now if you remember in the beginning, there is a dtype warning says that columns 6 has mixed types
# If you count the column from zero in the first row, the 6th row is the Police Dept Code

# Now, let's examine Police Dept Code:

accidents["Police Dept Code"].unique()

# We do this to see what are the possible unique inputs in Police Dept Code
```

```python
# Look at these, there are '1' and 1, a mixed type between integer and strings

# Now, in order to solve this, we need to change the data type in the column in to strings
# We change it by using function .astype to change it's data type

accidents["Police Dept Code"] = accidents["Police Dept Code"].astype(str)

# Now, let's examine if the Police Dept Code has changed into string type

accidents["Police Dept Code"].dtype
```

```python
# Examine the data inside Police Dept Code to be sure
# We do this by using unique function again

accidents["Police Dept Code"].unique()

# Yes, there is no integer anymore! They are all strings!
```

```python
# Let's find out the unique value (output) for column Alcohol Involved

accidents["Alcohol Involved"].unique()
```

```python
# Ok, now we want to see how many N and how many Y in alcohol involved columns
# We do this by using fucntion .value_counts()

# We can get the output without using pd.DataFrame, but we use it to make the output looks nicer in a frame

alcohol = pd.DataFrame(accidents["Alcohol Involved"].value_counts())
alcohol 
```

```python
# Now, let's find if there is any null or NA values in alcohol involved column
# We do this by using .isnull() function followed by .sum() fucntion to sum up the total of null values in the column

accidents["Alcohol Involved"].isnull().sum()
```

```python
# Now, with the concern of that the columns will contain nulls values and NA out of the way
# We proceed to calculate what is the percentage of accidents are alcohol involved

alcohol["Percent"] = (alcohol["Alcohol Involved"]/accidents["Alcohol Involved"].count()) * 100
alcohol

# Notice we uses alcohol, because alcohol is in a new DataFrame
# Previously, the DataFrame only have Alcohol Involved column, now we added one more new column called Percent by
# Writing alcohol["Percent"]
```

```python
# Due to some confusion while writing the formula, this are the difference between .value_counts() and .count() function
# .value_counts[] count the value in each unique data

accidents["Alcohol Involved"].value_counts()
```

```python
# While .count() just count the total data

accidents["Alcohol Involved"].count()
```

# 3:

```python
# Let's see how many accidents happen in each county

accidents["County Name"].value_counts()
```

```python
# Ok, what we see above is just visuals, now we need to group them up by county

accidents.groupby("County Name")

# Now, we have successfully group them up by county
```

```python
# After we grouped each accidents by its county name 
# (it just mean they are grouped by county but still will have mutiple rows for each accidents inputs)
# We have to sum up those accidents that is group by county then only we can observe the summary

county_accidents = accidents.groupby("County Name").sum()
pd.DataFrame(county_accidents)
```

```python
# Ok, now we want to take out Total Killed and put it into a new dataframe

county_death = accidents.groupby("County Name").sum().iloc[ : , 0].sort_values(ascending = False)

# iloc[ : , 0] means integer location[rows, columns] in our case it will be [: , 0]
# meaning, for every row, take the first column

# Of course, now we want to DataFrame it now

pd.DataFrame(county_death)
```

# 4:

```python
# Ok, now we want to work with dates
# First we need to import datetime from datetime in order for us to work with date

from datetime import datetime

accidents
```

```python
# Now, we apply datetime.strptime to all data Crash Date column with the format "%m/%d/%y"
# This is to turn every string data type into the date format "%m/%d/%y"

accidents["Crash Date"]= accidents["Crash Date"].apply(lambda x: datetime.strptime(x, "%m/%d/%y").date())
```

```python
# Now, let's see the converted data into date time

accidents["Crash Date"]
```

```python
# Now, let group them by crash date and count them

crash_by_date = accidents.groupby("Crash Date").count().iloc[:,0]
crash_by_date
```

```python
# Now, let's sort it

crash_by_date.sort_values(ascending = False)

# 2008-02-12 has the most accidents
```

```python
# Now, let's use bokeh to visualise data

from bokeh.plotting import figure, show, output_file, output_notebook
from bokeh.models import HoverTool
```

```python
# We sort the total death by accident in order

sorted_death = county_death_bar.sort_values(by = "Total Killed", ascending = False)
sorted_death
```

```python
# Then, use bokeh package to visualise our data

output_notebook()

county_name = sorted_death["County Name"]

bar = figure(title = "Total Killed by County",
             x_range = county_name,
             plot_width = 800,
             plot_height = 600,
             toolbar_location = None,
             tools = "")

bar.vbar(x = "County Name",
         top = "Total Killed",
         source = sorted_death, # This is where you put your data
         width = 0.8)

bar.xaxis.major_label_orientation = "vertical"
bar.xgrid.grid_line_color = None
bar.y_range.start = 0

output_file("your_first_bokeh_barplot.html")

show(bar)
```

```python
# Now, let's use bokeh package to work with scatter plot

county_accidents_scatter
```

```python
# This is how you use bokeh to make scatter plot

scatterplot = figure(title = "The Relationships between Total Killed and Pedestrians Killed in Every County",
                     y_axis_label = "Pedestrians Killed",
                     x_axis_label = "Total Killed")

scatterplot.circle("Total Killed",
                   "Pedestrians Killed",
                   source = county_accidents_scatter) # This is where you input your data

output_file("scatterplot.html")

show(scatterplot)
```

```python
# With the scatter plot, now we can make a regression line
# To do that, we need seaborn package
# Let's import it

import seaborn as sns
```

```python
# Using package function .lmplot we can plot the regression line

sns.lmplot(x= "Total Killed", y= "Pedestrians Killed", data = county_accidents_scatter) 
```

```python
# We import matplotlib to use plt function to zoom in on the range of our regression line

import matplotlib.pyplot as plt

sns.jointplot(x = "Total Killed",
              y = "Pedestrians Killed",
              data = county_accidents_scatter,
              kind = "reg",
              joint_kws = {"color":"green"})

# Use these two function to zoom in into the range of our regression line to fit the chart
# plt.ylim(0, 70)
# plt.xlim(0, 180)
```
