# 2:

```python
# Let's say if you want to make a new column
# You put: file_name["New_column"] = 

census_data["black_alone_percentage"] = (census_data["black_alone"]/census_data["total_population"])*100
census_data.head()
```

```python
# If you want to sort the data
# Let's say you want to sort it by values, so sort_values
# sort_values(by= "condition")
# Then you can add on, ascending = False => big to small

census_data.sort_values(by = "median_income", ascending = False) # sort_values() sorts in ascending order
```

```python
### Groupby

# You can group data using .groupby(["key"])["key_values"]
# Then you can add on function to it, for our case, we added on .median() 
# To find median black_alone_percentage median that is groupby county

census_data.groupby(["county"])["black_alone_percentage"].median()
```

```python
## Other than the method show above, we have other method => Aggregate (New vocab)

# Now on the left hand side:
# We want it to .groupby(["key"])

# Then we .agg() function
# Inside it we use dictionary {} because we using "sum"

# Then .agg({"data": "method"})
# Basically in our case, we find total_population and sum it up, then sort it by state


census_data.groupby(["state"]).agg({"total_population": "sum"})
```

```python
# Import new data

data = pd.read_csv("countries.csv")
data

# We can't see all the countries data, as there are repeated country data for all the year
# So how do we solve this issue?
# We use frequency table
```

```python
# Frequency table can show you how many times a country appear

frequency_table = pd.crosstab(index = data["country"], columns= "counts")

# pd.crosstab => Compute a simple cross table of two (or more) factors.
# In our case is index(rows) and columns

# So what happen here is we use pd.crosstab to compute cross tab with some factors:
# So pd.crosstab(index = data_file["What data"], columns = "Naming")

# So basically across the table, we are finding how many index occur in data_file["data"], which means rows
# And we put that index number into frequency table columns named under counts

frequency_table
```

```python
# Now, we want to check if there is China in the list
# So we find by: data_file[data_file.column_name == "What we finding"]
# == is to find is there any equals to "What we finding"

# In our case, we are finding China, so is under column country, therefore:
data[data.country == "China"]
```

```python
# Now we want to plot the population of China and us
# We need to import matplotlib.pyplot

import matplotlib.pyplot as plt

# We need to define China and US
china = data[data.country == "China"]
us = data[data.country == "United States"]

plt.plot(china.year, china.population, label = "China", color = "red")
plt.plot(china.year, china.population, "o", color = "red")
plt.plot(us.year, us.population, label = "US", color = "blue")
plt.plot(us.year, us.population, "o", color = "blue")
plt.legend()
plt.title("Population Each Year")
plt.xlabel("Year")
plt.ylabel("Population")
plt.show()

# Each point in the line is the population for that year
# Now what if we want to plot the growth??? How to do?
```

```python
# Let's see the data for china population
china.population
```

```python
# In order to find the growth, we have to divide the population of that year over the population in first year
# We use iloc, means integer location to locate the first year population

china.population / china.population.iloc[0] * 100
```

```python
# Now simply plot the graph

plt.plot(china.year, china.population/china.population.iloc[0] * 100, label = "China", color = "red")
plt.plot(china.year, china.population/china.population.iloc[0] * 100, "o", color = "red")
plt.plot(us.year, us.population/us.population.iloc[0] * 100, label = "US", color = "blue")
plt.plot(us.year, us.population/us.population.iloc[0] * 100, "o", color = "blue")
plt.legend()
plt.title("Population Growth Rate Comparisons between China and the US")
plt.xlabel("Year")
plt.ylabel("Population Growth\n(The First Year in the Data is set at 100)")
plt.show()
```
